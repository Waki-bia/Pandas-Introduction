```python
## select categorical variables
tips_cat = tips_data.select_dtypes(exclude='number')
tips_cat.head()
```

```python
## value counts for each of the categorical variable
[tips_cat[name].value_counts() for name in tips_cat]
```

```python
## Get proportion for each of categories
## value counts for each of the categorical variable
[tips_cat[name].value_counts(normalize=True) for name in tips_cat]
```

```python
## select the continuous variables
df_cont = tips_data.drop(columns=tips_cat)
df_cont
```

```python
## find the correlations of the dataset
df_cont.corr()
```

```python
## Correlation of all variables with price
df_cont.corrwith(df_cont['price']).abs().sort_values(ascending=False)
```

## Pivot Tables and Cross-Tabulation

### Pivot Tables

```python
## load the data
tips_data.head()
```

```python
### Generate a pivot table using rows with average as the aggregate function
tips_data.pivot_table(values = 'price', columns=['bimas', 'varieties'],
aggfunc=np.mean).unstack()
```

```python
## counts of the variables
tips_data[['status', 'bimas']].value_counts().unstack()
```

```python
## Proportions expressed in a contigency table
## counts of the variables
tips_data[['status', 'bimas']].value_counts(normalize = True).unstack()
```

```python
## Using different columns and rows
tips_data.pivot_table(values='price', index = 'bimas', columns='status', aggfunc=np.sum, margins=True)
```

```python
## Use two different indexes
tips_data.pivot_table(values='price', index = ['bimas', 'varieties'], columns='status', aggfunc=np.sum,
## fill Na with zero
fill_value=0
)
```

```python
## Different representation of the same with wide format
## Use two different indexes
tips_data.pivot_table(values='price', index = ['bimas', 'varieties'], columns='status', aggfunc=np.sum,
## fill Na with zero
fill_value=0
).unstack()
```

### Cross-Tabulation

```python
## Do a cross-tabulation for different categorical variables
pd.crosstab(tips_data['varieties'], tips_data['status'])
```

```python
## Proportion of the crosstabs
pd.crosstab(tips_data['varieties'], tips_data['status'], normalize=True)
```

```python
## Cross-tabs with marginal tabs
pd.crosstab(tips_data['varieties'], tips_data['status'], margins=True)
```
